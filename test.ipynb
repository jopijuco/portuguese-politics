```python
import requests

# data updated daily
PATH = "https://app.parlamento.pt/webutils/docs/doc.txt?path=6148523063446f764c324679626d56304c3239775a57356b595852684c3052685a47397a51574a6c636e52766379394a626d6c6a6157463061585a68637939595356596c4d6a424d5a57647063327868644856795953394a626d6c6a6157463061585a686331684a566c3971633239754c6e523464413d3d&fich=IniciativasXIV_json.txt&Inline=true"

payload = requests.get(PATH)
assert payload.status_code == 200
```

```python
iniciativas = payload.json().get("ArrayOfPt_gov_ar_objectos_iniciativas_DetalhePesquisaIniciativasOut", {}).get("pt_gov_ar_objectos_iniciativas_DetalhePesquisaIniciativasOut", {})
```

```python
def to_list(x):
  return x if isinstance(x, list) else [x]

a = []
for d in iniciativas:
    events = to_list(d.get("iniEventos", {}).get("pt_gov_ar_objectos_iniciativas_EventosOut", []))
    for event in events:
        votacao = event.get("votacao", {}).get("pt_gov_ar_objectos_VotacaoOut", {})
        if (isinstance(votacao, list)):
          votacao = votacao[0]
        a.append(votacao.get("detalhe", "").lower().replace(" ", "").replace("<br>", "").replace("</i>", "").replace("<i>", "").replace("afavor:", ""))

        v = votacao.get("detalhe", "").lower().replace(" ", "").replace("<br>", "").replace("</i>", "").replace("<i>", "")
        if votacao and not v.startswith("afavor:"):
          print(v)
        
```
